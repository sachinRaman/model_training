```python
import numpy as np
import pandas as pd
```

```python
input_data_df = pd.read_csv('./chat_dataset.csv')
```

```python
input_data_df.head()
input_data_df.describe()
```

```python
input_data_df
```

```python
dialogues = input_data_df['dialogue'].tolist()
SEP = '<>'
```

```python
dialogues
```

```python
dialogues_list = []

for dialogue in dialogues:
    conv_list = dialogue.split(SEP)[:-1]
    new_conv_list = []
    for conv in conv_list:
        new_conv_list.append(conv.strip())
    dialogues_list.append(new_conv_list)
```

```python
dialogues_list
```

```python
final_data_list = []
for conversation in dialogues_list:
    i=1
    conv = ''
    for dialogue in conversation:
        if i%2 == 1:
            conv += 'Human: ' + dialogue + '\nBot: '
        else:
            final_data_list.append([conv, dialogue])
            conv += dialogue +'\n'
        i += 1
```

```python
final_data_list
```

```python
final_df = pd.DataFrame(final_data_list, columns=['input', 'output'])
```

```python
final_df
```

```python
# Calculate the sum of words for each row
final_df['word_count'] = final_df['input'].str.split().str.len() + final_df['output'].str.split().str.len()

# Remove rows where the sum of words is greater than 1200
final_df = final_df[final_df['word_count'] <= 1200]

# Remove the 'word_count' column if desired
final_df = final_df.drop('word_count', axis=1)

# Print the resulting DataFrame
print(final_df)
```
